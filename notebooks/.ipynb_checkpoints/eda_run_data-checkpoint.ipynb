```python
# --- Stage 1A: Data Quality Audit ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Load data
data_path = Path("../data/strava/processed/strava_runs.parquet")
df = pd.read_parquet(data_path)
print(f"✅ Loaded {len(df):,} rows × {len(df.columns)} columns")
```

```python
# --- 1. Basic summary ---
display(df.head())
display(df.info())
display(df.describe(include='all').T)
```

```python
# --- 2. Missingness overview ---
missing = df.isna().mean().sort_values(ascending=False)
display(missing)

plt.figure(figsize=(10,6))
missing.head(20).plot(kind='barh', title='Top 20 Columns by Missing %')
plt.gca().invert_yaxis()
plt.show()

# --- 3. Quick visual checks ---
sns.histplot(df["distance_km"], bins=50, kde=True)
plt.title("Distribution of distance_km"); plt.show()

sns.histplot(df["pace_min_per_km"], bins=50, kde=True)
plt.title("Distribution of pace_min_per_km"); plt.show()
```

```python
# Group by run_id to see completeness per run
run_summary = (
    df.groupby("run_id")
      .agg(
          records=('timestamp', 'count'),
          total_distance_km=('distance_km', 'max'),
          avg_pace=('pace_min_per_km', 'mean'),
          cadence_mean=('cadence', 'mean'),
          hr_mean=('heart_rate', 'mean'),
          elevation_max=('altitude', 'max'),
      )
      .reset_index()
)

print(run_summary.describe())
sns.histplot(run_summary["total_distance_km"], bins=60, kde=True)
plt.title("Run distance distribution"); plt.show()

sns.scatterplot(data=run_summary, x="total_distance_km", y="avg_pace")
plt.title("Distance vs average pace"); plt.show()
```

```python
# --- earlier cells ---
# Load data
# Basic summary (df.info, describe)
# Missingness plot
# Run-level completeness plots

# --- ADD THIS: Data Quality Manifest ---
import numpy as np
import pandas as pd
from pathlib import Path

print("🔍 Generating Data Quality Manifest...")

manifest = pd.DataFrame({
    "non_null_count": df.notna().sum(),
    "total_rows": len(df),
})
manifest["completeness_%"] = (manifest["non_null_count"] / manifest["total_rows"]) * 100
manifest["dtype"] = df.dtypes.astype(str)

def classify_column(col, completeness):
    col_lower = col.lower()
    if "unknown" in col_lower:
        return "drop"
    if col_lower in [
        "distance", "distance_km", "pace_min_per_km", "timestamp",
        "speed", "cadence", "altitude", "enhanced_speed",
        "lat", "lon", "position_lat", "position_long", "run_id"
    ]:
        return "core"
    if completeness < 20:
        return "drop"
    if completeness < 60:
        return "optional"
    return "core"

manifest["category"] = [
    classify_column(col, completeness)
    for col, completeness in manifest["completeness_%"].items()
]

manifest = manifest.sort_values("completeness_%", ascending=False)
manifest_path = Path("../data/strava/processed/data_quality_manifest.csv")
manifest.to_csv(manifest_path, index_label="column")

print(f"✅ Saved data quality manifest → {manifest_path}")
display(manifest.head(15))
```
